```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
import plotly.express as px 
```

```python
df = pd.read_csv('positions.csv')
df
```

```python
df = df.drop(['name', 'p', 'w', 'd', 'l', 'f', 'a', 'gd'], axis = 1)
df.reset_index(inplace = True, drop = True)
df.season = df.season.str[:4]

df['season'] = pd.to_datetime(df['season'])
df['season'] = df['season'].dt.year
df = df.sort_values(by = ['season','team'])
df[:50]
```

```python
df[:44].sort_values(by = 'team')
```

```python
df['team'].value_counts()
```

```python
fig = px.scatter(df[22:], x = 'points', y = 'pos', animation_frame='season', animation_group = 'team', size='pos', hover_name='team', log_x=True, size_max=45, color = 'team')
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 700
fig.show()
```
